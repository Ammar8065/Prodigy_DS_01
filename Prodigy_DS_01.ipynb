```python
#Importing Libraries
import pandas as pd
import matplotlib.pyplot as plt
```

```python
#Loading the data
file1 = (r"D:/Data science/New/API_SP.POP.TOTL_DS2_en_csv_v2_462652.csv")
file2 = (r"D:/Data science/New/Metadata_Country_API_SP.POP.TOTL_DS2_en_csv_v2_462652.csv")
file3 = (r"D:/Data science/New/Metadata_Indicator_API_SP.POP.TOTL_DS2_en_csv_v2_462652.csv")
```

```python
# Reading the CSV files
data1 = pd.read_csv(file1, skiprows=4)  # Skipping the first 4 rows which often contain metadata
data2 = pd.read_csv(file2)
data3 = pd.read_csv(file3)
```

```python
# Extracting relevant columns from the population data
population_data = data1[['Country Name', 'Country Code', '2023']].dropna()
population_data = population_data.rename(columns={'2023': 'Population_2023'})
```

```python
# Sorting data by population for better visualization
population_data = population_data.sort_values(by='Population_2023', ascending=False)
```

```python
#Selecting the top 30 countries
top_population_data = population_data.head(30)
```

```python
# Creating a bar chart for the top 20 countries by population in 2023
plt.figure(figsize=(14, 8))
plt.bar(top_population_data['Country Name'], top_population_data['Population_2023'], color='skyblue')
plt.xlabel('Country')
plt.ylabel('Population in 2023')
plt.title('Top 20 Countries by Population in 2023')
plt.xticks(rotation=90)
plt.tight_layout()
plt.show()
```
